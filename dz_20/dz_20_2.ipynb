#### Програма сортування з dz_13_2 із застосуванням магічної команди %%timeit

```python
"""Програма сортування списку"""

from random_words import RandomWords


str_list = []
w = RandomWords()
for i in range(0, 5000):
    str_list.append(w.random_word())
print("List for sorting:", str_list)


def bubble_sort(data):
    """Bubble Sort Algorithm"""
    length = len(data)
    for i_index in range(length):
        swapped = False
        for j_index in range(0, length - i_index - 1):
            if data[j_index] > data[j_index + 1]:
                data[j_index], data[j_index + 1] = data[j_index + 1], data[j_index]
                swapped = True
        if not swapped:
            break
    return data


def sort(lst):
    bubble_sort(lst)
    print("After sorting: ", lst)


sort(str_list)
```

```python
%%timeit -n 1 -r 5
import time
time.sleep(1)
```
